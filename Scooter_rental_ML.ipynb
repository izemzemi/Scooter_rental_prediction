```python
#Bar charts
def PlotBarCharts(inpData, colsToPlot):
    
    # Generating multiple subplots
    fig, subPlot=plt.subplots(nrows=1, ncols=len(colsToPlot), figsize=(20,5))
    fig.suptitle('Bar charts of: '+ str(colsToPlot))

    for colName, plotNumber in zip(colsToPlot, range(len(colsToPlot))):
        inpData.groupby(colName).size().plot(kind='bar',ax=subPlot[plotNumber])
```

```python
PlotBarCharts(inpData=scoot_rental, colsToPlot=[
    'holiday', 'year', 'month', 'day', 'hour', 'weather'])
```

```python
import seaborn as sns
#Create the correlation matrix
correMtr=scoot_rental[['Temp','humidity','wind','holiday','year','month','day','hour','unknown','subscriber','n']].corr()
mask=np.array(correMtr)
mask[np.tril_indices_from(mask)]=False
#Heat map for correlation matrix of attributes
fig,ax=plt.subplots(figsize=(15,8))
sns.heatmap(correMtr,mask=mask,vmax=0.8,square=True,annot=True,ax=ax)
ax.set_title('Correlation matrix of attributes')
plt.show()
```

#### We note that the unkonwn count is more correlated to the  weather variables (humidity, wind, temperature). This is expected as subscribers use the scooters more often and are less likely to be deterred by uncomfortable weather conditions, contrarily to unkown users (more likely to be tourists or using the scooters for entertainment purpose). In general, the number of rented scooters  show a positive correlation with temperature and a negative correlation with humidity. Also, we notice that some of the time variables have a high correlation with the bicycle counts; this is especially true for hours. However, this is less true for the day. Since the total count of the rented scooters 'n', unkown and subscriber  are highly positively correlated to each other. So, we are going to ignore unkown and subscriber variables in the  following analysis.


```python
#load the required libraries
from sklearn import preprocessing,metrics,linear_model
from sklearn.model_selection import cross_val_score,cross_val_predict,train_test_split
```

```python
#Convert categorial features to category type
scoot_rental.holiday=scoot_rental.holiday.astype('category')
scoot_rental.year=scoot_rental.year.astype('category')
scoot_rental.weather=scoot_rental.weather.astype('category')

#Choosing the data frame on which we will operate the machine learning
df_train =scoot_rental[["weather","holiday","Temp","humidity","wind","year","month","day","hour","n"]]
```

```python
#Split the dataset into the train and test data
X_train,X_test,y_train,y_test=train_test_split(df_train.iloc[:,0:-1],df_train.iloc[:,-1],test_size=0.3, random_state=42)

#Reset train index values
X_train.reset_index(inplace=True)
y_train=y_train.reset_index()

# Reset train index values
X_test.reset_index(inplace=True)
y_test=y_test.reset_index()
#Shape verification
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
print(y_train.head())
print(y_test.head())
```

```python
#categorical attributes
cat_attributes=['weather','holiday','year']
#numerical attributes
num_attributes=['Temp','humidity','wind','month','day','hour']
```

```python
#To get dummy variables to encode the categorical features to numeric
X_train_encoded=pd.get_dummies(X_train,columns=cat_attributes)
print('Shape of transfomed dataframe::',X_train_encoded.shape)
X_train_encoded.head(5)
```

```python
#Training dataset for modelling
X_train=X_train_encoded
y_train=y_train.n.values
```

## Machine learning models using scikit-learn

### Linear regression method

```python
#training model
lr_model=linear_model.LinearRegression()
lr_model
```

```python
#fit the trained model
lr_model.fit(X_train,y_train)
```

```python
#Accuracy of the model
lr=lr_model.score(X_train,y_train)
print('Accuracy of the model :',lr)
print('Model coefficients :',lr_model.coef_)
print('Model intercept value :',lr_model.intercept_)
```

```python
#Cross validation prediction
predict=cross_val_predict(lr_model,X_train,y_train,cv=3)
predict
```

```python
#Cross validation plot
fig,ax=plt.subplots(figsize=(15,8))
ax.scatter(y_train,y_train-predict)
ax.axhline(lw=2,color='black')
ax.set_title('Cross validation prediction plot')
ax.set_xlabel('Observed')
ax.set_ylabel('Residual')
plt.show()
```

```python
#Cross validation plot
fig,ax=plt.subplots(figsize=(15,8))
ax.plot(y_train[0:100])
ax.plot(predict[0:100])
#ax.axhline(lw=2,color='black')
ax.set_title('Cross validation prediction plot')
#ax.set_xlabel('Observed')
ax.set_ylabel('y')
plt.show()
```

```python
#R-squared scores
r2_scores = cross_val_score(lr_model, X_train, y_train, cv=3)
print('R-squared scores :',np.average(r2_scores))
```

```python
#To get dummy variables to encode the categorical features to numeric
X_test_encoded=pd.get_dummies(X_test,columns=cat_attributes)
print('Shape of transformed dataframe :',X_test_encoded.shape)
X_test_encoded.head(5)
```

```python
#Test dataset for prediction
X_test=X_test_encoded
y_test=y_test.n.values
```

### Random forest

```python
#Training the model
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=200)
#Fit the trained model
rf.fit(X_train,y_train)
#accuracy of the model
rf_score =rf.score(X_train,y_train)
print('Accuracy of the model :',rf_score)
#Cross validation prediction
predict=cross_val_predict(rf,X_train,y_train,cv=3)
predict
```

```python
#Cross validation prediction plot
fig,ax=plt.subplots(figsize=(15,8))
ax.scatter(y_train,y_train-predict)
ax.axhline(lw=2,color='black')
ax.set_title('Cross validation prediction plot')
ax.set_xlabel('Observed')
ax.set_ylabel('Residual')
plt.show()
```

```python
#R-squared scores
r2_scores = cross_val_score(rf, X_train, y_train, cv=3)
print('R-squared scores :',np.average(r2_scores))
#predict the model
rf_pred=rf.predict(X_test)
rf_pred
```

```python
#Root mean square error
rmse = math.sqrt(metrics.mean_squared_error(y_test,rf_pred))
print('Root mean square error :',rmse)
#Mean absolute error
mae=metrics.mean_absolute_error(y_test,rf_pred)
print('Mean absolute error :',mae)
```

```python
#Residual scatter plot
fig, ax = plt.subplots(figsize=(15,8))
residuals=y_test-rf_pred
ax.scatter(y_test, residuals)
ax.axhline(lw=2,color='black')
ax.set_xlabel('Observed')
ax.set_ylabel('Residuals')
ax.set_title('Residual plot')
plt.show()
```
